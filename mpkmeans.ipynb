```python
import numpy as np
from kmeans import StandardKMeans1, StandardKMeans2, mp1KMeans, mp2KMeans,  allowKMeans1,  allowKMeans2, chop
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

LOW_PREC = chop(np.float16)
```

```python
np.random.seed(0)
x, _ = make_blobs(n_samples=5000, n_features=100, centers=3000, random_state=36)
x = (x - x.mean(axis=0)) / x.std(axis=0)
```

```python
kmeans1 = StandardKMeans1(n_clusters=5, seeding='d2', random_state=0, verbose=1)
kmeans1.fit(x)
```

```python
kmeans2 = StandardKMeans1(n_clusters=5, seeding='d2', random_state=0, verbose=1)
kmeans2.fit(x)
```

```python
mpkmeans1 = mp1KMeans(n_clusters=5, seeding='d2', low_prec=LOW_PREC, random_state=0, verbose=1)
mpkmeans1.fit(x)

```

```python
mpkmeans2 = mp2KMeans(n_clusters=5, seeding='d2', low_prec=LOW_PREC, random_state=0, verbose=1)
mpkmeans2.fit(x)

```

```python
alkmeans1 = allowKMeans1(n_clusters=5, seeding='d2', low_prec=LOW_PREC, random_state=0, verbose=1)
alkmeans1.fit(x)
```

```python
alkmeans2 = allowKMeans2(n_clusters=5, seeding='d2', low_prec=LOW_PREC, random_state=0, verbose=1)
alkmeans2.fit(x)

```

```python
plt.semilogy(kmeans1.inertia, label='float64 (dist1)',marker='*', linestyle='-', markersize=5)
plt.semilogy(kmeans2.inertia, label='float64 (dist2)',marker='*', linestyle='--', markersize=3)
plt.semilogy(mpkmeans1.inertia, label='mp1' ,marker='v')
plt.semilogy(mpkmeans2.inertia, label='mp2' ,marker='+')
plt.semilogy(alkmeans1.inertia, label='all low precision 1' ,marker='v')
plt.semilogy(alkmeans2.inertia, label='all low precision 2' ,marker='+')
plt.legend()
```
